#### Next, evaluate if the columns in this dataset are strongly correlated.

If our dataset has strongly correlated columns, we need to choose certain ML algorithms instead of others. We need to evaluate this for our dataset now.

Luckily, most of the columns in this dataset are ordinal which makes things a lot easier for us. In the next cells below, evaluate the level of collinearity of the data.

We provide some general directions for you to consult in order to complete this step:

1. You will create a correlation matrix using the numeric columns in the dataset.

1. Create a heatmap using `seaborn` to visualize which columns have high collinearity.

```python
# Your code here
#we # Step 1: Select only numeric columns for correlation analysis
numeric_cols =websites.select_dtypes(include =[np.number]).columns.tolist()
print("Numeric columns:", numeric_cols )

#2 : compute the correlation matrix
corr_matrix = websites[numeric_cols].corr()

# 3 Create the heatmap
plt.figure(figsize=(14,20))
sns.heatmap(
    corr_matrix,
    annot = True,   #show the correlation valiues inside each cell
    fmt ='.2f',
    cmap='coolwarm',  # red=positive, blue=negative correlation
    vmin=-1, vmax=1,  # correlation always between -1 and 1
    linewidths=0.5
)
plt.title("Correlation Heatmap - Website Features", fontsize=14)
plt.tight_layout()
plt.show()

# Show pairs with high collinearity (>0.90)
print("\nHighly correlated pairs (|corr| > 0.90):")
for i in corr_matrix.columns:
    for j in corr_matrix.columns:
        if i < j and abs(corr_matrix.loc[i, j]) > 0.90:
            print(f"  {i} <-> {j}: {corr_matrix.loc[i, j]:.3f}")
```

```python
# Your comment here
# Your comment here
# Observations from the heatmap:
# HIGH COLLINEARITY GROUPS (|corr| > 0.90):
# 1. URL_LENGTH <-> NUMBER_SPECIAL_CHARACTERS (0.92) → keep URL_LENGTH, drop NUMBER_SPECIAL_CHARACTERS
# 2. TCP_CONVERSATION_EXCHANGE <-> SOURCE_APP_PACKETS <-> REMOTE_APP_PACKETS <-> APP_PACKETS (~0.99)
#    → keep TCP_CONVERSATION_EXCHANGE, drop the other 3
# 3. APP_BYTES <-> REMOTE_APP_BYTES (1.00 — perfect!) → keep APP_BYTES, drop REMOTE_APP_BYTES
# Strategy: remove the minimum number of columns to break collinearity groups
print("Identified 5 columns to remove due to high collinearity.")
```

Not yet. If you reflect on the previous labs how we handle categorical columns, you probably remember we ended up dropping a lot of those columns because there are too many unique values. Too many unique values in a column is not desirable in machine learning because it makes prediction inaccurate. But there are workarounds under certain conditions. One of the fixable conditions is:

#### If a limited number of values account for the majority of data, we can retain these top values and re-label all other rare values.

The `WHOIS_COUNTRY` column happens to be this case. You can verify it by print a bar chart of the `value_counts` in the next cell to verify:

```python
# Your comment here
# SERVER has many variants but they all fall into 3 main categories:
# - Microsoft (IIS, ASP.NET, etc.)
# - Apache (various versions)
# - nginx (various versions)
# All others → 'Other'
# Strategy: check if each value CONTAINS 'Microsoft', 'Apache', or 'nginx' as a substring
print("Plan: consolidate SERVER into 4 categories: Apache, nginx, Microsoft, Other")
```

Although there are so many unique values in the `SERVER` column, there are actually only 3 main server types: `Microsoft`, `Apache`, and `nginx`. Just check if each `SERVER` value contains any of those server types and re-label them. For `SERVER` values that don't contain any of those substrings, label with `Other`.

At the end, your `SERVER` column should only contain 4 unique values: `Microsoft`, `Apache`, `nginx`, and `Other`.

```python
# Your code here
def classify_server(server_name):
    """Classify server into 4 main categories based on substring matching."""
    s = str(server_name).lower()  # lowercase for case-insensitive matching
    if 'microsoft' in s:
        return 'Microsoft'
    elif 'apache' in s:
        return 'Apache'
    elif 'nginx' in s:
        return 'nginx' 
    else:
        return 'Other'

websites['SERVER'] = websites['SERVER'].apply(classify_server)

print("SERVER value counts after consolidation:")
print(websites['SERVER'].value_counts())
```

```python
# Count `SERVER` value counts here
print("Final SERVER distribution:")
print(websites['SERVER'].value_counts()) 
```

# Challenge 6 - Modeling, Prediction, and Evaluation

We'll start off this section by splitting the data to train and test. **Name your 4 variables `X_train`, `X_test`, `y_train`, and `y_test`. Select 80% of the data for training and 20% for testing.**

```python
from sklearn.model_selection import train_test_split

# Your code here:

X = website_dummy.drop(columns=['Type'])   # features — everything except the target
y = website_dummy['Type']                  # target — what we want to predict (0=benign, 1=malicious)

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,     # 20% for testing, 80% for training
    random_state=42    # fixed seed for reproducibility
)

print(f"X_train: {X_train.shape}  |  y_train: {y_train.shape}")
print(f"X_test:  {X_test.shape}   |  y_test:  {y_test.shape}")
```

#### In this lab, we will try two different models and compare our results.

The first model we will use in this lab is logistic regression. We have previously learned about logistic regression as a classification algorithm. In the cell below, load `LogisticRegression` from scikit-learn and initialize the model.

```python
# Your code here:
from sklearn.linear_model import LogisticRegression
# Initialize: create an empty (untrained) model# crear un mdeolo vacui sin entrnar 
lr_model = LogisticRegression(max_iter=2000, random_state=42)
print("Logistic Regression model initialized:", lr_model)
```

Next, fit the model to our training data. We have already separated our data into 4 parts. Use those in your model.

```python
# Your code here:
#Next , fit the model to aour training data
lr_model.fit(X_train, y_train) #X_train : practicas examen #y_train: respuetas del ex
print("✓ Logistic Regression trained!")
print(f"  Training score: {lr_model.score(X_train, y_train):.4f}")

```

finally, import `confusion_matrix` and `accuracy_score` from `sklearn.metrics` and fit our testing data. Assign the fitted data to `y_pred` and print the confusion matrix as well as the accuracy score

```python
# Your code here:

# Your code here:
from sklearn.metrics import confusion_matrix, accuracy_score

# Generate predictions on TEST data (data the model has NEVER seen)
y_pred_lr = lr_model.predict(X_test)

# Confusion Matrix
cm_lr = confusion_matrix(y_test, y_pred_lr)
print("=== Logistic Regression ===")
print("Confusion Matrix:")
print(cm_lr)
print(f"  True Negatives  (benign  correctly identified): {cm_lr[0,0]}")
print(f"  False Positives (benign  called malicious):     {cm_lr[0,1]}")
print(f"  False Negatives (malicious missed):             {cm_lr[1,0]}")
print(f"  True Positives  (malicious correctly caught):   {cm_lr[1,1]}")
print(f"\nAccuracy Score: {accuracy_score(y_test, y_pred_lr):.4f} ({accuracy_score(y_test, y_pred_lr)*100:.2f}%)")
```

What are your thoughts on the performance of the model? Write your conclusions below.

```python
#### Your conclusions here:
cm_lr = confusion_matrix(y_test, y_pred_lr)
print(cm_lr)
#Training test score
train_acc_lr = lr_model.score(X_train, y_train)
# Final test score
test_acc_lr  = accuracy_score(y_test, y_pred_lr) #y_pred_lr = lr_model.predict(X_test)

print("=== Logistic Regression Performance Analysis ===")
print(f"Training Accuracy: {train_acc_lr:.4f} ({train_acc_lr*100:.2f}%)")
print(f"Test Accuracy:     {test_acc_lr:.4f}  ({test_acc_lr*100:.2f}%)")
print(f"Gap (train-test):  {(train_acc_lr - test_acc_lr):.4f}")
print()
print("Conclusions:")
print("- Accuracy ~95%: the model is very good at distinguishing benign from malicious websites.")
print("- Train ≈ Test: small gap means no significant overfitting — model generalizes well.")
print("- The model misses some malicious sites (False Negatives) — in cybersecurity, this is the critical error.")
print("- Class imbalance (88% benign) means accuracy alone can be misleading.")
print("  A model that always predicts 'benign' would get ~88% accuracy but catch ZERO malicious sites!")
```

```python
# Your code here

# Your code here
from sklearn.preprocessing import RobustScaler

# Step 1: Initialize the scaler
scaler = RobustScaler()

# Step 2: Fit on TRAINING data ONLY, then transform both train and test
# IMPORTANT: never fit on test data — that would be data leakage!
X_train_scaled = scaler.fit_transform(X_train)  # fit + transform training data
X_test_scaled  = scaler.transform(X_test)        # transform only (use training stats)

print("Scaling applied successfully!")
print(f"X_train_scaled shape: {X_train_scaled.shape}")
print(f"X_test_scaled shape:  {X_test_scaled.shape}")

# Step 3: Train Logistic Regression on scaled data
lr_scaled = LogisticRegression(max_iter=2000, random_state=42)
lr_scaled.fit(X_train_scaled, y_train)

# Step 4: Predict and evaluate
y_pred_scaled = lr_scaled.predict(X_test_scaled)
cm_scaled = confusion_matrix(y_test, y_pred_scaled)
acc_scaled = accuracy_score(y_test, y_pred_scaled)

print("\n=== Logistic Regression WITH RobustScaler ===")
print("Confusion Matrix:")
print(cm_scaled)
print(f"\nAccuracy: {acc_scaled:.4f} ({acc_scaled*100:.2f}%)")

print("\n=== Final Comparison ===")
print(f"LR without scaling: {test_acc_lr:.4f}  ({test_acc_lr*100:.2f}%)")
print(f"LR with scaling:    {acc_scaled:.4f}  ({acc_scaled*100:.2f}%)")
diff = acc_scaled - test_acc_lr
print(f"Change: {'+' if diff >= 0 else ''}{diff:.4f}")
if abs(diff) < 0.005:
    print("Minimal difference — Logistic Regression is fairly robust to scaling for this dataset.")
elif diff > 0:
    print("Improvement! Feature scaling helped Logistic Regression converge better.")
else:
    print("No improvement — the scaled model performed similarly or slightly worse.")
```
